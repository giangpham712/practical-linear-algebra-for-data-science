## Visualization

```python
plt.plot(1, 2, 'ko')
plt.plot([0, 2], [0, 4], 'r--')
plt.xlim([-4, 4])
plt.ylim([-4, 4])
plt.title('The graph title')
```

```python
_, axs = plt.subplots(1, 2, figsize=(8, 5))
print(np.random.randn(10, 5))
axs[0].plot(np.random.randn(10, 5))
axs[1].imshow(np.random.randn(10, 5))
```

```python
x = np.linspace(-4, 4, 42)
y = x ** 2

plt.plot(x, y, 's-')
```

```python
x = np.linspace(-4, 4, 42)
alpha = 1.4
beta = 2

num = alpha
den = 1 + np.exp(-beta * x)
fx = num / den

plt.plot(x, fx, 's-')
```
